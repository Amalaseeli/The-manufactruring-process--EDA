```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
 
```

```python
df=pd.read_csv('../failure_data.csv')
```

```python
df
```

```python
df=df.drop('Unnamed: 0', axis=1)
```

```python
df
```

```python
df.describe()
```

```python
plt.figure(figsize=(20,10))
df.hist(bins=50)
plt.show()
```

```python
df.convert_dtypes()
```

```python
df.dtypes
```

```python
columns_to_boolean=['Machine failure','TWF','HDF', 'PWF', 'OSF','RNF']
for col in columns_to_boolean:
    df[col]=df[col].astype('bool')

df.dtypes
```

```python
df['Type']=df['Type'].astype('category')
df.dtypes
```

```python
plt.figure(figsize=(12,8))
df.hist(bins=50 )
plt.subplots_adjust(hspace=0.8, wspace=0.7)
plt.tight_layout()
plt.show()
```

```python
numeric_columns=['UDI','Air temperature [K]','Process temperature [K]','Rotational speed [rpm]','Torque [Nm]', 'Tool wear [min]']
for col in numeric_columns:
    print(f"{col}:", df[col].skew())
```

```python
numeric_columns=['UDI','Air temperature [K]','Process temperature [K]','Rotational speed [rpm]','Torque [Nm]', 'Tool wear [min]']
for col in numeric_columns:
    
    sns.boxplot(df[col])
    plt.tight_layout()
    plt.show()
```
